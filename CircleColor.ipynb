```python
from google.colab.patches import cv2_imshow
import cv2
import numpy as np
```

```python
# Task of changing circle color through masking

image = cv2.imread("abc.png")
print("Original Image:")
cv2_imshow(image)

# By setting values in BGR plane
image2 = image
lower_bound = np.array([0, 0, 0])
upper_bound = np.array([220, 220, 255])
mask1 = cv2.inRange(image, lower_bound, upper_bound)
image2[mask1 > 0] = [0, 0, 0]
print("Upgraded Image by altering BGR values:")
cv2_imshow(image2)

# By setting HSV values
image3 = image
hsv_image = cv2.cvtColor(image3, cv2.COLOR_BGR2HSV)
lower_bound = np.array([0, 0, 0])
upper_bound = np.array([200, 100, 100])
mask2 = cv2.inRange(hsv_image, lower_bound, upper_bound)
image3[mask2 > 0] = [0, 0, 0]
print("Upgraded Image by altering HSV values:")
cv2_imshow(image3)




```
